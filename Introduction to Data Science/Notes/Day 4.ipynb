#  Accessing Values in a DataFrame 

```python
my_df = pd.DataFrame(np.random.randint(0,100,(100,5)))

my_df
```

# loc

```python
my_df.loc[0,0]      # Syntax - loc[row,column]
                    # this will return element present on 0th row and 0th column
```

```python
my_df.loc[4,0]      # will return element present on 4th row and 0th column
```

```python
my_df.columns = list("ABCDE")     # Renaming column names

my_df
```

```python
my_df.loc[0,'A']       # will return 0th element from 'A'th column
```

```python
my_df.loc[[1,2],['B','C','D']]   # passing multiple rows and columns also like this to access dataFrame.
                                 # will return elements from [1st and 2nd row] of [B'th, C'th and D'th column]
```

```python
my_df.loc[:,['B','C']]     # using concept of slicing to access elements.
                           # will return all elements of ['B' and C' column] 
```

```python
my_df.loc[[1,2],:]    # will return all elements of row 1 and row 2
```

```python
my_df.loc[ ( my_df['A']<20 ) ]   # We can apply conditions also so as to obtain desired output.
                                 # will return the whole dataFrame but, column 'A' will only be having elements which are less 
                                 # than 20.
```

```python
my_df.loc[(my_df['A']<5) & (my_df['D']>50)]   # using bitwise 'and' operator 
```

```python
my_df.loc[(my_df['A']<5) | (my_df['D']>90)]  # using bitwse 'or' operator
```

# iloc

```python
my_df
```

```python
my_df.iloc[0,4]   # iloc[] is integer index based, so we have to specify row index value and column index value 
                  # Syntax - iloc[row_index, column_index]
                  # Notice, here we are not specifying the name of the coulmn 'A', 'B' or 'C', instead we are using index value 
                  # of that column.
                
```

```python
my_df.iloc[[0,1],[3,4]]
```

```python
my_df.iloc[[0,1],:]
```

```python
my_df.iloc[:,[1,2]]
```

# Some Useful methods while working with DataFrame

```python
my_df.loc[:,'B'].isnull()   # isnull() method is used to check if there are any null elements or not
                            # if there are any null elements isnull() will return True otherwise False.
                            # In this case there is no null element in the column 'B'
```

```python
my_df.loc[:,'B']=None    # assigning 'none' to all elements of column 'B' 

my_df
```

```python
my_df.loc[:,'B'].isnull()  # returning 'True' on all elements in column 'B' which are = None
```

```python
my_df.isnull()  # applying isnull() on whole dataframe returns the dataframe with value "True" if the element is null 
                # and "False" if the element is not null.
```

```python
my_df.shape    # return the shape of dataFrame - (rows and columms)
```

```python
my_df
```

```python
my_df['C'].value_counts()    # Return a Series containing counts of unique rows in the DataFrame.
```

```python
my_df.notnull()   # Just opposite of .isnull() function.
```

```python
my_df.mean()   # Will return the mean value of each column 
```

```python
my_df.corr()   # Used to find the pairwise correlation of all columns in the dataframe. 
               # Any na values are automatically excluded.
    
               # Advance concept
```

```python
my_df.count()  # returns the count of total elements from each column
```

```python
my_df.max()    # returns the maximum value from each column
```

```python
my_df.min()    # returns the minimum value from each column
```

```python
my_df.std()   # returns the standard deviation of each column
```

```python
my_df.drop('B',axis=1)   # drop() method drops the/removes the specified labels from rows or columns.
                         # In this case column 'B' has been dropped.
                         # Since by default axis is = 0,  so we need to specify axis = 1.
```

```python
marvel = {
            "Name":[None, "Tony", "Banner", "Tony","Tony",None, "Thor"],
            "Dates":[None, "12-2-12", "10-4-11", "6-7-12", "4-5-6",None, "2-2-2"]
         }
df_marvel = pd.DataFrame(marvel)
df_marvel
```

```python
df_marvel.drop_duplicates(subset=["Name"])   # will drop the duplicate values from "Name" and keeps the very first record
```

```python
df_marvel.drop_duplicates(subset=["Name"],keep='last') # will keep the very last record after removal of duplicates
```

```python
df_marvel.dropna()
```
